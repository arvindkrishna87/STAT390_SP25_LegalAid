```python
def bar_graph(cat, y_label='Num Of Repeats'):
    plt.figure(figsize=(12, 8))
    ax=sns.barplot(cat, color='#f73c1b')
    for p in ax.patches:
        height = p.get_height()
        if not np.isnan(height) and height > 0:
            ax.text(
            p.get_x() + p.get_width() / 2,
            height,
            f'{height:.1f}',
            ha='center',
            va='bottom',
            fontsize=14  # ğŸ‘ˆ Smaller font size for numbers
        )

    # Improve x-axis labels
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right', fontsize=14)
    ax.set_ylabel(y_label, fontsize=14)
    plt.tight_layout()
    plt.show()
```

```python
#this function iterates through a list, and checks whether a value is a repeat and if so, records the value in front of it
#this seeks to find flows/activities that cause users to go back to things they have already been to
def find_repeat_cause_flows(l):

    #this stores the number of times each value appears in the list, used to identify repeats
    counter = flow_dict.copy()

    #this stores the number of times each value "causes" a repeat
    repeat_counter = flow_dict.copy()
    
    for i in range(len(l)):
        counter[l[i]]+=1
        if counter[l[i]]>1:
            repeat_counter[l[i-1]]+=1
    #gets rid of zeroes in the dictionary
    return {key: value for key, value in repeat_counter.items() if value != 0}      
```

```python
#this function iterates through a list and find the number of times each value appears back to back
def find_repeat_activities(l):
    #used to store number of b2b repeats
    counter = act_dict.copy()
    for i in range(1,len(l)):

        #checks if item is the same as the one before it
        if l[i]==l[i-1]:
            counter[l[i]]+=1

    #gets rid of zeroes
    return {key: value for key, value in counter.items() if value != 0}
```

```python
def find_between_repeats(col, activity=True, limit=0): #limit is for the minimum number of instances to not be classified as other
    if activity:
        journey=act_journey
        counts=activity_counts
    else:
        journey=flow_journey
        counts=flow_counts
    
    #filters the activity_counts series to only show calls that repeat a certain activity/flow
    repeated=counts.apply(lambda x: x[col]>=2) 
    
    #for the ones where senior repeat, this finds the indices in the journey where the caller goes to the seniors menu
    where_repeats=journey[repeated[repeated==True].index].apply(lambda x: [i for i, val in enumerate(x) if val==col])

    #filters for only non-consecutive repeats
    repeats_distanced=where_repeats[where_repeats.apply(lambda x: x[-1]-x[0]>len(x)-1)]

    #made dataframe with contact session IDs, their activity journey, and their difference from above, made a df for the next apply function to be easier
    df=pd.concat([journey[repeats_distanced.index],repeats_distanced], axis=1)
    df.rename(columns={0:'journey',1:'indices'}, inplace=True)
    
    #find the activities in the journey that fall between the first and last repeated activities

    if activity:
        between_repeats=df.apply(lambda x: Counter(x['journey'][x['indices'][0]+1:x['indices'][-1]]),axis=1)
        between_count=act_dict.copy()
        for x in between_repeats:
            for key in x.keys(): 
                between_count[key]+=x[key]
        between_count={key: value for key, value in between_count.items() if value != 0}
        between_count=pd.Series(between_count).sort_values(ascending=False)
    else:
        between_repeats=df.apply(lambda x: x['journey'][x['indices'][0]+1:x['indices'][-1]],axis=1)
        between_count=between_repeats.value_counts()

    #make graph readable
    condensed_between = between_count[between_count>=limit]
    condensed_between_count=pd.concat([condensed_between,pd.Series(between_count[between_count<limit].sum(),
                                                                                     index=['Other'])])
    if activity:  
        bar_graph(condensed_between_count, 'Num of Instances Between Repeats')
    print('Percent of callers who land on', col, 'that repeat:', round(between_repeats.shape[0]/journey[journey.apply(lambda x: col in x)].shape[0]*100,2),'%')
    print(between_repeats.shape[0])
    print(journey[journey.apply(lambda x: col in x)].shape[0])


    return between_repeats
```

```python
def has_non_consecutive_duplicate(lst, skip_values=None):
    if skip_values is None:
        skip_values = set()
    else:
        skip_values = set(skip_values)

    seen = {}
    for i, val in enumerate(lst):
        if val in skip_values:
            continue  
        if val in seen:
            if i - seen[val] > 1:
                return True
        seen[val] = i
    return False

```

## Non-consec-summary

```python
from collections import defaultdict

def count_non_consecutive_repeats(lst):
    positions = defaultdict(list)
    
    for i, val in enumerate(lst):
        positions[val].append(i)
    
    result = {}
    for val, idxs in positions.items():
        # Check if there are at least two indices
        if len(idxs) > 1:
            # Check for at least one non-consecutive repeat
            non_consec = any(abs(a - b) > 1 for a, b in zip(idxs, idxs[1:]))
            if non_consec:
                result[val] = sum(abs(a - b) > 1 for a, b in zip(idxs, idxs[1:]))
    
    return result
```

```python
non_consec_repeats_all_activities = act_journey.apply(count_non_consecutive_repeats)
non_consec_act_count = act_dict.copy()
for x in non_consec_repeats_all_activities:
    for key in x.keys(): 
        non_consec_act_count[key]+=1
non_consec_act_count = {key: value for key, value in non_consec_act_count.items() if value != 0}
non_consec_act_count=pd.Series(non_consec_act_count).sort_values(ascending=False)
non_consec_act_count.drop(['PlayMOH300s','QueueMenu1'], inplace=True)
condensed_non_consec_act_count = non_consec_act_count[non_consec_act_count>=1000]
condensed_non_consec_act_count=pd.concat([condensed_non_consec_act_count,pd.Series(non_consec_act_count[non_consec_act_count<1000].sum(),
                                                                                     index=['Other'])])
bar_graph(condensed_non_consec_act_count, 'Num Of Callers Who Repeat')
```

```python
non_consec_repeats_all_flows = flow_journey.apply(count_non_consecutive_repeats)
non_consec_flow_count = flow_dict.copy()
for x in non_consec_repeats_all_flows:
    for key in x.keys(): 
        non_consec_flow_count[key]+=1
non_consec_flow_count = {key: value for key, value in non_consec_flow_count.items() if value != 0}
non_consec_flow_count=pd.Series(non_consec_flow_count).sort_values(ascending=False)
condensed_non_consec_flow_count = non_consec_flow_count[non_consec_flow_count>=200]
condensed_non_consec_flow_count=pd.concat([condensed_non_consec_flow_count,pd.Series(non_consec_flow_count[non_consec_flow_count<150].sum(),
                                                                                     index=['Other'])])
bar_graph(condensed_non_consec_flow_count, 'Num of Callers Who Repeat')
```
