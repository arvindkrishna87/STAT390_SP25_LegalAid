```python
def fix_LACMain(x):
    if len(x)>1:
        if x[1] == 'LACMain':
            x[1] = 'MainMenu'
    return x
```

```python
def fix_LACMain2(x):
    if len(x)>1:
        if x[1] == 'LACMain':
            x[1] = 'HolidayClosedMenu'
    return x
```

```python
#this function iterates through a list, and checks whether a value is a repeat and if so, records the value in front of it
#this seeks to find flows/activities that cause users to go back to things they have already been to
def find_repeat_cause_flows(l):

    #this stores the number of times each value appears in the list, used to identify repeats
    counter = flow_dict.copy()

    #this stores the number of times each value "causes" a repeat
    repeat_counter = flow_dict.copy()
    
    for i in range(len(l)):
        counter[l[i]]+=1
        if counter[l[i]]>1:
            repeat_counter[l[i-1]]+=1
    #gets rid of zeroes in the dictionary
    return {key: value for key, value in repeat_counter.items() if value != 0}      
```

```python
#above function applied on the series
repeat_flow_causes=flow_journey.apply(lambda x: find_repeat_cause_flows(x))
```

```python
#aggregates the series to get a total count for each flow
total_flow_cause_count=flow_dict.copy()
for x in repeat_flow_causes:
    for key in x.keys(): 
        total_flow_cause_count[key]+=x[key]

#removes the dict entries that have a value of zero
total_flow_cause_count={key: value for key, value in total_flow_cause_count.items() if value != 0}

#sorts and converts to a series
total_flow_cause_count=pd.Series(total_flow_cause_count).sort_values(ascending=False)
```

```python
#graph
plt.figure(figsize=(10, 6))
ax=sns.barplot(total_flow_cause_count, color='#f73c1b')
for p in ax.patches:
    height = p.get_height()
    if not np.isnan(height) and height > 0:
        ax.text(
            p.get_x() + p.get_width() / 2,
            height,
            f'{height:.1f}',
            ha='center',
            va='bottom',
            fontsize=8  # 👈 Smaller font size for numbers
        )

ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
#this function iterates through a list and find the number of times each value appears back to back
def find_repeat_activities(l):
    #used to store number of b2b repeats
    counter = act_dict.copy()
    for i in range(1,len(l)):

        #checks if item is the same as the one before it
        if l[i]==l[i-1]:
            counter[l[i]]+=1

    #gets rid of zeroes
    return {key: value for key, value in counter.items() if value != 0}
```
