```python
# cleaning function

def clean(df):
    df.columns = df.iloc[1]
    new = df.iloc[2:].reset_index(drop = True)

    new.replace('N/A', pd.NA, inplace=True)
    new.fillna(method='ffill', inplace=True)
    return new

```

```python
# cleaning dfs

xjan25_1 = clean(jan25_1)
xjan25_2 = clean(jan25_2)
xfeb25_1 = clean(feb25_1)
xfeb25_2 = clean(feb25_2)
xmar25_1 = clean(mar25_1)
xapr24_1 = clean(apr24_1)
xapr24_2 = clean(apr24_2)
xmay24_1 = clean(may24_1)
xmay24_2 = clean(may24_2)
xjun24_1 = clean(jun24_1)
xjun24_2 = clean(jun24_2)
xjul24_1 = clean(jul24_1)
xjul24_2 = clean(jul24_2)
xaug24_1 = clean(aug24_1)
xaug24_2 = clean(aug24_2)
xsep24_1 = clean(sep24_1)
xsep24_2 = clean(sep24_2)
xoct24_1 = clean(oct24_1)
xoct24_2 = clean(oct24_2)
xnov24_1 = clean(nov24_1)
xnov24_2 = clean(nov24_2)
xnov24_3 = clean(nov24_3)
xdec24_1 = clean(dec24_1)
xdec24_2 = clean(dec24_2)
xdec24_3 = clean(dec24_3)
```

```python
# list of cleaned dfs

all_data = [xjan25_1 ,xjan25_2 ,xfeb25_1 ,xfeb25_2 ,xmar25_1 ,xapr24_1 ,xapr24_2 ,xmay24_1 ,xmay24_2 ,xjun24_1 ,xjun24_2 
,xjul24_1, xjul24_2 ,xaug24_1 ,xaug24_2 ,xsep24_1 ,xsep24_2 ,xoct24_1 ,xoct24_2 ,xnov24_1 ,xnov24_2 
,xnov24_3 ,xdec24_1 ,xdec24_2 ,xdec24_3 ]

len(all_data)
```

```python
# concatenating the list of clean dfs 
combined = pd.DataFrame(columns=xjan25_1.columns)

for dataset in all_data:
    combined = pd.concat([combined, dataset[:len(dataset) - 1]], axis = 0)
```
